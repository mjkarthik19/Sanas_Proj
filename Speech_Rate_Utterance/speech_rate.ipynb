```python
with open(r'C:\Users\JanardhanMuralidhar\Sanas_Projects\Speech_Rate_Utterance\Transcripts\transcript1.json') as file:
    transcript_data = json.load(file)

# Function to count syllables in a word
def count_syllables(word):
    # Load the CMU Pronouncing Dictionary
    cmudict = nltk.corpus.cmudict.dict()

    # Convert the word to lowercase
    word = word.lower()

    # Check if the word is in the CMU Pronouncing Dictionary
    if word in cmudict:
        # Get the list of phonemes for the word
        phonemes = cmudict[word][0]

        # Count the number of phonemes that end with a digit
        syllable_count = len([ph for ph in phonemes if ph[-1].isdigit()])
    else:
        # If the word is not found in the dictionary, assign a default count of 1
        syllable_count = 1

    return syllable_count

total_duration = 0
words = transcript_data['channels']['alternatives'][0]['words']
# Iterate over the words in the transcript and calculate syllables per second
total_syllables = 0

for word_data in words:
    word = word_data['word']
    start_time = word_data['start']
    end_time = word_data['end']
    word_duration = end_time - start_time

    syllables = count_syllables(word)
    syllables_per_second = syllables / word_duration

    total_syllables += syllables
    total_duration += word_duration

    print(f"Word: {word}")
    print(f"Syllables: {syllables}")
    print(f"Duration: {word_duration} seconds")
    print(f"Syllables per Second: {syllables_per_second}")
    print()

# Calculate the SPM metric
spm = total_syllables / total_duration
print(f"Speech rate (SPM): {spm}")
```

```python
import json

def extract_sentences(transcript_file):
    with open(transcript_file, 'r') as file:
        transcript_data = json.load(file)

    sentences = []
    utterances = transcript_data.get('utterances', [])

    for utterance in utterances:
        sentence = utterance.get('transcript', '')
        sentences.append(sentence)

    return sentences

transcript_file = r'C:\Users\JanardhanMuralidhar\Sanas_Projects\Speech_Rate_Utterance\Transcripts\transcript5.json'
sentences = extract_sentences(transcript_file)
print(sentences)
```

```python
import json
import nltk

cmudict = nltk.corpus.cmudict.dict()

def count_syllables(word):
    word = word.lower()
    if word in cmudict:
        phonemes = cmudict[word][0]
        syllable_count = len([ph for ph in phonemes if ph[-1].isdigit()])
    else:
        syllable_count = 1
    return syllable_count

def calculate_syllables_per_second(transcript_file):
    with open(transcript_file, 'r') as file:
        transcript_data = json.load(file)

    utterances = transcript_data['utterances']

    total_syllables = 0
    total_duration = 0

    for utterance in utterances:
        sentence = utterance['transcript']
        start_time = utterance['start']
        end_time = utterance['end']
        duration = end_time - start_time

        words = utterance['words']
        syllables = sum(count_syllables(word['word']) for word in words)
        total_syllables += syllables
        total_duration += duration

        syllables_per_second = syllables / duration

        # Print the results for each utterance
        print(f"Utterance: {sentence}")
        print(f"Syllables: {syllables}")
        print(f"Duration: {duration} seconds")
        print(f"Syllables per Second: {syllables_per_second}")
        print()

    # Print the total syllables, total duration, and syllables per second for the transcript
    print(f"Total Syllables: {total_syllables}")
    print(f"Total Duration: {total_duration} seconds")
    print(f"Syllables per Second: {total_syllables / total_duration}")
```
