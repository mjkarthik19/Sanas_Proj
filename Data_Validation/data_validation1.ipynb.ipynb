```python
import re
import editdistance

def find_matching_words(column1_text, column2_text):
    column1_words = re.findall(r"\b\w+(?:[.,!?']\w+)*\b|[.,!?<>]", column1_text)
    column2_words = re.findall(r"\b\w+(?:[.,!?']\w+)*\b|[.,!?<>]", column2_text)
    matching_words = []

    for word in column1_words:
        best_match = None
        best_distance = float('inf')
        for column2_word in column2_words:
            distance = editdistance.eval(word.lower(), column2_word.lower())
            if distance < best_distance:
                best_distance = distance
                best_match = column2_word

        if best_match is not None:
            matching_words.append(best_match)

    return matching_words


def remove_extra_characters(word):
    cleaned_word = re.sub(r"(^[^\w.,!?']+|[^\w.,!?']+$)", "", word)
    return cleaned_word

extracted_data = []

# Iterate over the data
for i in range(len(column1)):
    column1_text = column1[i]
    column2_text = column2[i]

    # Find the matching words in column2
    matching_words = find_matching_words(column1_text, column2_text)

    # Remove extra characters from matching words
    cleaned_matching_words = [remove_extra_spaces(word) for word in matching_words]

    # Join the cleaned matching words to form the final extracted text
    matching_text = " ".join(cleaned_matching_words)

    # Add the extracted data to the list
    extracted_data.append((column1_text, matching_text))

# Storing the extracted matching data in a variable
output = ""

# Printing the extracted matching data
for column1_text, matching_text in extracted_data:
    output += f"Column 1: {column1_text}\n"
    output += f"Column 2: {matching_text}\n"
    output += "\n"

# Print the accumulated output
print(output)
```
